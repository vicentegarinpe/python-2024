1.- generar un vector aleatorio que incluya numeros negativos desde el -10 hasta el 10 positivo.este vector debe tener 25 elementos. deben ser numeros flotantes

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
vector= np.random.uniform(-10,10,25)
vector
```

transformar el vector en una matriz cuadrada de 5x5 obtener el promedio de la matriz

```python
matriz= np.reshape(vector,(5,5))
matriz
promedio=np.mean(matriz)
```

generar un mapa de calor a partir de la matriz anterior

```python
#crear una figura y un eje
fig,grafic= plt.subplots()

#muestra la imagen del mapa de calor de la matriz
csgrafico= grafic.imshow(matriz,cmap="hot",interpolation="nearest")

fig.colorbar(csgrafico)


```
